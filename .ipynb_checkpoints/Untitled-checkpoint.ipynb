```python
import pandas as pd
import numpy as np
```

```python
price_file = "../Price/FB.csv"
price = pd.read_csv(price_file, usecols=['Date', 'Adj Close'])
```

```python
sec_file = "../SEC_8K/Data/8K data.h5"
sec = pd.read_hdf(sec_file)
```

```python
twitter_file = "../Twitter/twitter.h5"
twitter = pd.read_hdf(twitter_file)
```

```python
price = price.set_index("Date")
price.head()
```

```python
col = {}
for i in range(50):
    col[i] = "sec{}".format(i)
```

```python
sec = pd.DataFrame(sec.values.tolist(), index=sec.index).rename(columns=col)
sec.head()
```

```python
twitter.head()
```

```python
data = price.join(twitter).dropna().join(sec).fillna(0)
```

```python
data['next_day_value'] = data['Adj Close'].shift(-14)
```

```python
data = data.dropna()
```

```python
data.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data_1 = data.iloc[:1000, :]
data_2 = data.iloc[1000:, :]
```

```python
print(data_1.shape)
print(data_2.shape)
```

```python
train, test = train_test_split(data_1, test_size=0.3, shuffle=False)
train = train.copy()
test = test.copy()
```

```python
train = train.append(data_2)
```

```python
scaler = MinMaxScaler(feature_range=(-1,1))
scaler.fit(train["Adj Close"].values.reshape(-1,1))
test["Adj Close"] = scaler.transform(test["Adj Close"].values.reshape(-1,1))
test["next_day_value"] = scaler.transform(test["next_day_value"].values.reshape(-1,1))
train["Adj Close"] = scaler.transform(train["Adj Close"].values.reshape(-1,1))
train["next_day_value"] = scaler.transform(train["next_day_value"].values.reshape(-1,1))
```

```python
def create_data(df, days, features):
    num_data = df.shape[0]
    df2 = df.values
    sequences = num_data - days + 1
    x = np.zeros((sequences, days*features))
    y = np.zeros((sequences))
    dates = []
    for seq in range(sequences):
        x[seq, :] = df2[seq:seq+days, 0:features].reshape(days*features)
        y[seq] = df2[seq+days-1, features]
        dates.append(df.index[seq+days-1])
    return x,  y, dates
```

```python
x_train, y_train, _ = create_data(train, 40, 52)
x_test, y_test, dates = create_data(test, 40, 52)
```

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.optimizers import Adam
from keras import regularizers
from keras.callbacks import History 
import math
import matplotlib.pyplot as plt
```

```python
model = Sequential()
model.add(Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01), input_shape=(x_train.shape[1],)))
model.add(Dropout(0.9))
model.add(Dense(1))
model.compile(loss='mean_squared_error',
             optimizer='adadelta',
             metrics=['mse'])
```

```python
history = History()
model.fit(x_train, y_train, epochs=60, batch_size=56, callbacks=[history], shuffle=False, validation_split=0.05)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
mse, _ = model.evaluate(x_test, y_test)
print("LSTM Test MSE: {}".format(mse))
print("LSTM Test RMSE: {}".format(math.sqrt(mse)))
```

```python
model.save_weights("../two_week_ann_new.h5")
```
